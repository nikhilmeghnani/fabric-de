```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
spark.sql("""
CREATE TABLE IF NOT EXISTS earth_lakehouse_2.dim_customer (
    customer_id STRING,
    customer_name STRING,
    address STRING,
    city STRING,
    state STRING,
    zip_code STRING,
    start_date DATE,
    end_date DATE,
    is_active BOOLEAN
)
USING DELTA;
""")

spark.sql("""
CREATE TABLE IF NOT EXISTS earth_lakehouse_2.stg_customer (
    customer_id STRING,
    customer_name STRING,
    address STRING,
    city STRING,
    state STRING,
    zip_code STRING
)
USING DELTA;
""")
```

```python
spark.sql("""
MERGE INTO earth_lakehouse_2.dim_customer AS target
USING earth_lakehouse_2.stg_customer AS source
ON target.customer_id = source.customer_id 
AND target.is_active = TRUE  -- Match only active records

WHEN MATCHED AND (
    target.address <> source.address OR 
    target.city <> source.city OR 
    target.state <> source.state OR 
    target.zip_code <> source.zip_code
) 
THEN 
-- Close the old record
UPDATE SET target.end_date = CURRENT_DATE(), target.is_active = FALSE 

WHEN NOT MATCHED 
THEN 
-- Insert new updated record
INSERT (customer_id, customer_name, address, city, state, zip_code, start_date, end_date, is_active)
VALUES (source.customer_id, source.customer_name, source.address, source.city, source.state, source.zip_code, CURRENT_DATE(), NULL, TRUE);

""")
```

```python
spark.sql("""
SELECT * FROM earth_lakehouse_2.dim_customer ORDER BY customer_id, start_date;
""").show()
```
